```python
# YOUR CODE HERE
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import GridSearchCV
```

<b>Task:</b> Use the rest of this notebook to carry out your project plan. 

You will:

1. Prepare your data for your model.

Add code cells below and populate the notebook with commentary, code, analyses, results, and figures as you see fit. 

```python
# YOUR CODE HERE
# to remove observations with NA values
df = df.dropna()
# to select features that are more relevant to life ladder
features = df.corr()['Life Ladder'].abs().sort_values(ascending=False).drop('Life Ladder')
features = features[:10,].index.tolist()
```

```python
# define features and the label
X = df[features]
y = df['Life Ladder']
```

```python
# divide training and test set, with test set being 0.25 of all observations
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=1234)
```

```python
# construct and fit the model
# Initialize the model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)

# Calculate R-squared
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

Since for this model, I used the default linear regression model settings, I'll use grid search to find if there's any parameter that builds a better model that avoids overfitting.

```python
param_grid = {
    'alpha': [0.01, 0.1, 1, 10, 100],  # Different values of alpha to test
    'solver': ['auto', 'svd', 'cholesky', 'lsqr', 'saga']  # Different solvers
}
```

```python
# Initialize the model
ridge = Ridge()

# Set up the GridSearchCV
grid_search = GridSearchCV(estimator=ridge, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit the grid search to the data
grid_search.fit(X_train, y_train)
```

```python
# Best parameters from the grid search
print("Best parameters:", grid_search.best_params_)

# Best score achieved
print("Best score (negative MSE):", grid_search.best_score_)
```

```python
model_best = Ridge(alpha=0.1, solver='lsqr')
model_best.fit(X_train,y_train)
y_predict = model_best.predict(X_test)
```

```python
# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_predict)

# Calculate R-squared
r2 = r2_score(y_test, y_predict)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

In this case, the linear regression may be better covering the variances of the data.
